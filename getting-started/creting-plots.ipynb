# Criando gráficos

O dataset air_quality_no2.csv possui os valores de NO2 das estações de medidas FR04014, BETR801 e London Westerminister, localizadas respectivamente em Paris, Antwerp e Londres.

```python
# Importando dataset csv
import urllib.request
url_dataset = 'https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv'
dataset_csv = './data/air_quality_no2.csv'
urllib.request.urlretrieve(url_dataset, dataset_csv);
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_quality = pd.read_csv('./data/air_quality_no2.csv', index_col=0, parse_dates=True)
air_quality.head()
```

```python
air_quality.plot();
```

```python
# Apenas as colunas com dados de Paris
air_quality['station_paris'].plot();
```

```python
# Comparando valores de NO2 medidos em Londres x Paris
air_quality.plot.scatter(x='station_london',y='station_paris',alpha=0.5);
```

```python
# Visão geral dos métodos gráficos disponíveis
[
    method_name
    for method_name in dir(air_quality.plot)
    if not method_name.startswith("_")
]
```

```python
air_quality.plot.box();
```

```python
# Colunas separadas em diferentes subplots
axs = air_quality.plot.area(figsize=(12,4),subplots=True);
```

```python
# Customizando e salvando o gráfico resultante
fig, axs = plt.subplots(figsize=(12,4)) # Cria figura(Figure) e eixos (Axes) do matplotlib vazios
air_quality.plot.area(ax=axs) # Usa Pandas para colocar a área do gráfico na Figure/Axes preparados
axs.set_ylabel('N0$_2$ concentration') # Realiza customizações com o matplot lib (aqui está alterando o rótulo do eixo y)
fig.savefig("./outputs/no2_concentrations.png") # Salva a Figure/Axes usando um método existente na biblioteca matplotlib
```
