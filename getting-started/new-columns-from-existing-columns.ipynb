# Criando novas colunas derivadas de colunas existentes

```python
# Importando dataset csv
#import urllib.request
#url_dataset = 'https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/air_quality_no2.csv'
dataset_csv = './data/air_quality_no2.csv'
#urllib.request.urlretrieve(url_dataset, dataset_csv);
```

```python
import pandas as pd
```

```python
air_quality = pd.read_csv(dataset_csv)
air_quality.head()
```

```python
#Convertento a concentração de NO2 da estação em Londres em mg/m³ e criando uma nova coluna
# Considerando temperatura de 25°C e pressão de 1013hPa, o fator de conversão é 1882 
air_quality['london_mg_per_cubic'] = air_quality['station_london']*1.882
air_quality.head()
```

```python
# Verificando a razão dos valores em Paris versus Antwerpe salvando o resultado em uma nova coluna
air_quality['ratio_paris_antwerp'] = (air_quality['station_paris']/air_quality['station_antwerp'])
air_quality.head()
```

```python
# Renomeando os dados das colunas com identificadores da estação corespondente usados por openAQ
air_quality_renamed = air_quality.rename(
    columns={
        "station_antwerp": "BETR801",
        "station_paris": "FR04014",
        "station_london": "London Westminster",
    }
)

air_quality_renamed.head()
```

```python
# Renomeando as colunas para lower case
air_quality_renamed = air_quality_renamed.rename(columns=str.lower)

air_quality_renamed.head()
```
